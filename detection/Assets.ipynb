### Update 360 image collection yay

```python
# Script for separating sites based on detections #
## Option 1: search for or provide a list of files by filename ##
from cognite.client import CogniteClient

def update_image_collection(file_names, new_site, old_site, client):
    for filename in file_names:
        res = client.files.search(name=filename, filter={"metadata": {"site_id": old_site}})
        if res:  # Check if the search returned any results
            file = res[0]  # Take only the first file from the results
            print(file.name)
    
            # Process only if the first file's site_id matches old_site
            if file.metadata.get("site_id") == old_site:
                station_id = file.metadata.get("station_id") #The image itself
                print("Processing station ID:", station_id)
    
                # Searching for files specifically at one station and site, and get the file id
                res = client.files.search(name="", filter={"metadata": {"station_id": station_id, "site_id": old_site}})
                file_ids = [file.id for file in res]  # Extracting file IDs
                print("Files at station", station_id, ":", [file.name for file in res])
                print("File IDs:", file_ids)
    
                #update site information for all respective files
                res = client.files.retrieve_multiple(ids=file_ids)
                for file in res:
                    file.metadata["site_id"] = new_site
                    file.metadata["site_name"] = new_site
                    client.files.update(file)
                print("Updated files' site information.")
    
                #update site information for all respective events
                res = client.events.search(description="", filter={"metadata": {"station_id": station_id}})
                for event in res:
                    event.metadata["site_id"] = new_site
                    event.metadata["site_name"] = new_site
                    client.events.update(event)
                print("Updated events' site information.")
        else:
            print(f"No results found for {filename}")
```
